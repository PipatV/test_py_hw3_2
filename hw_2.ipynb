#นำเข้าlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#อ่านไฟล์

```python
df = pd.read_csv("telecom_churn.csv")
```

#เลือก Churn 0 และ 1

```python
df_0 = df[df.Churn==0]
df_1 = df[df.Churn==1]
```

#รวมผลลัพธ์

```python
df_sum_0 = sum(df_0.MonthlyCharge)
df_sum_1 = sum(df_1.MonthlyCharge)
```

#ค่าเฉลี่ยของแต่ละกลุ่ม

```python
df_sum_00 = df_sum_0/(len(df_0))
df_sum_01 = df_sum_1/(len(df_1))
```

```python
print(df_sum_00)
print(df_sum_01)
```

#เลือกกลุ่ม

```python
plot_0 = df_0.MonthlyCharge
plot_1 = df_1.MonthlyCharge
```

```python
print(type(plot_0))
```

```python
print(plot_0)
```

#แปลงข้อมูลกลุ่ม 0เป็น dic เพื่อนำไป plot

```python
dict = plot_0.to_dict()
```

```python
dict_0_1 = dict.keys()
```

```python
dict_0_2 = dict.values()
```

#plot ของกลุ่ม0

```python
plt.title("การกระจายของกลุ่ม 0",fontname='Tahoma',fontsize=13)
plt.ylabel("การโทรต่อเดือนของกลุ่ม 0",fontname='Tahoma',fontsize=13)
plt.xlabel("ข้อมูลของกลุ่ม 0",fontname='Tahoma',fontsize=13)
plt.scatter(dict_0_1,dict_0_2)
plt.show()
```
